# Adding Delay to the FL technique

```python
fig,ax = plt.subplots(figsize = (8,8))
wpos = 1e5
dic ={"Side" :"Right","FF": True,"FFonset" : 0}
for delay in [0,0.06]:
    Feedback_Linearization(w1 = wpos,w2 = wpos,Num_iter = 60,Delay=delay,ShowEstimate=True,Activate_Noise=False,AdditionalDynamics=dic)

hand, labl = ax.get_legend_handles_labels()
handout,lablout =[],[]
for h,l in zip(hand,labl):
    if l not in lablout:
        lablout.append(l)
        handout.append(h)
fig.legend(handout, lablout)
plt.title("Impact of Delay")
plt.show()

fig,ax = plt.subplots(figsize = (8,8))

for delay in [0,0.06]:
    Feedback_Linearization(w1 = wpos,w2 = wpos,Num_iter = 60,Delay=delay,ShowEstimate=True,Activate_Noise=False,AdditionalDynamics=dic)

hand, labl = ax.get_legend_handles_labels()
handout,lablout =[],[]
for h,l in zip(hand,labl):
    if l not in lablout:
        lablout.append(l)
        handout.append(h)
fig.legend(handout, lablout)
plt.title("Impact of Delay")
plt.show()
```
